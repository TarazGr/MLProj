```python
with open("songs_1970_2018_uniq.csv", "r") as source:
    with open("songs_1970_2018.csv", "w", encoding="utf-8") as dest:
        lines = [line.rstrip('\n') for line in source]
        for index, line in enumerate(lines):
            if index == 0: 
                dest.write(line+'\n')
                continue
            month, day, year = line.split(',')[0].split('/')
            if len(year) != 4: 
                year = "20" + year if int(year) < 50 else "19" + year
            if len(day) == 1:
                day = '0' + day
            if len(month) == 1:
                month = '0' + month
            dest.write(year+'-'+month+'-'+day+',')
            for val in line.split(',')[1:]:
                dest.write(val+',')
            dest.write('\n')
```
